```python
from prophet import Prophet
```

```python
df = pd.read_csv("cafethai_prices.csv")
df['datetime'] = pd.to_datetime(df['date'] + " " + df['time'], dayfirst=True)
df['date'] = df['datetime'].dt.date
daily_sales = df.groupby(['date', 'coffee_name']).size().reset_index(name='cups_sold')
```

```python
# สมมติ daily_sales มีคอลัมน์: date, coffee_name, cups_sold
daily_sales['date'] = pd.to_datetime(daily_sales['date'], errors='coerce')
daily_sales.dropna(subset=['date'], inplace=True)

unique_drinks = daily_sales['coffee_name'].unique()
all_forecasts = []
```

```python
for drink in unique_drinks:
    # เลือกข้อมูลเครื่องดื่มแต่ละชนิด
    drink_data = daily_sales[daily_sales['coffee_name'] == drink][['date', 'cups_sold']].rename(columns={'date':'ds','cups_sold':'y'})

    # สร้างโมเดล Prophet
    model = Prophet(daily_seasonality=True, yearly_seasonality=True)
    model.fit(drink_data)

    # สร้าง future dataframe 60 วันล่วงหน้า
    future = model.make_future_dataframe(periods=60)
    forecast = model.predict(future)

    # เก็บเฉพาะ 60 วันล่วงหน้า
    future_forecast = forecast.tail(60)[['ds','yhat']]
    future_forecast['coffee_name'] = drink

    all_forecasts.append(future_forecast)
```

```python
# รวมข้อมูลทั้งหมด
result = pd.concat(all_forecasts)
result = result.rename(columns={'ds':'date','yhat':'predicted_cups'})
```

```python
# Pivot table ให้วันเป็นแถวและเมนูเป็นคอลัมน์
pivot_result = result.pivot(index='date', columns='coffee_name', values='predicted_cups')
print("ตาราง Pivot (จำนวนแก้วพยากรณ์ต่อวัน แบ่งตามเมนูเครื่องดื่ม):")
display(pivot_result)
```

```python
# Plot กราฟของแต่ละเครื่องดื่ม
for drink in unique_drinks:
    df_drink = result[result['coffee_name'] == drink]

    plt.figure(figsize=(12,4))
    plt.plot(df_drink['date'], df_drink['predicted_cups'])
    plt.title(f"60-Day Forecast of Cups Sold - {drink}")
    plt.xlabel("Date")
    plt.ylabel("Predicted Cups Sold")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
